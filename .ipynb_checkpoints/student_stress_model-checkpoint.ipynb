```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

<h1>Dataset import</h1>

```python
dataset = pd.read_csv("../datasets/student_dataset/Stress_Dataset.csv")
dataset.reset_index(drop=True, inplace=True)
dataset
```

<h3>Renaming Columns Name</h3>

```python
column_names = list(dataset.columns.unique())
column_names.remove('Have you been dealing with anxiety or tension recently?.1')
column_names
```

```python
new_column_name = [
    'Gender', 'Age', 'stress_in_life', 'rapid_heartbeat', 'anxiety_or_tension', 'sleep_problems', 'getting_headaches_often', 'irritated', 
    'focus_problem', 'sadness_or_low_mood', 'illness_or_health_issues', 'lonely_or_isolated', 'overwhelmed_workload', 'peer_pressure', 
    'relationship_problem', 'difficulty_with_professors', 'academic_environment_stressful', 'relaxation_trouble', 'environment_stressful',
    'lack_confidence_academic', 'lack_confidence_subjects', 'subject_conflicting', 'class_attendence', 'weight_gained', 'stress_level' 
]

new_dataset = dataset.drop(columns=['Have you been dealing with anxiety or tension recently?.1'])

new_dataset.columns = new_column_name

new_dataset.head()
```

<h1>Handling Missing Values</h1>

```python
new_dataset.fillna(0)
```

<h1>Making Dependent & Independent Variable</h1>

```python
X = new_dataset.iloc[:, :-1]
Y = new_dataset.iloc[:,-1]
```

<h3>Encoding Dependend Categorial Data</h3>

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
Y_encode = le.fit_transform(Y)
```

<h1>Spliting Data</h1>

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

<h1>Acuracy of the model</h1>

```python
accuracy_scores = []
k_values = range(1, 11)

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    accuracy_scores.append(acc)

best_k = k_values[accuracy_scores.index(max(accuracy_scores))]
cm = confusion_matrix(y_test, y_pred)
print(f"Best k = {best_k}, Accuracy = {max(accuracy_scores):.2f}")
print("Confusion Matrix:")
print(cm)
```

```python
plt.plot(k_values, accuracy_scores, marker='o')
plt.xlabel("k (Number of Neighbors)")
plt.ylabel("Accuracy")
plt.title("KNN Accuracy vs k")
plt.show()
```
